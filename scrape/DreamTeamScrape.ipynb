```python
def html_call(url):
    # here we define the headers for the request
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.13; rv:63.0) Gecko/20100101 Firefox/63.0'}

    # this request object will integrate your URL and the headers defined above
    req = urllib.request.Request(url=url, headers=headers)
    #req = urllib.request.Request(url=url, headers=headers)
    # calling urlopen this way will automatically handle closing the request
    
    time.sleep(0.5)
    with urllib.request.urlopen(req,timeout=5) as response:
        page_html = response.read()
    return page_html
```

```python
def fetch_dream_score(player_id):
    season_list = ['2100']
    scores_per_week = {}
    seasons = []
    weeks = []
    scores = []
    for season in range(21):
        # skip EM (3, 13) and WM (8, 18)
        if season in(2,12,7,17):
            continue
        print(f'season: {season+1}')
        season_str = f'{season+1}00'
        try:
            url = f'https://www.mydreamteam.ch/player_details.php?player={player_id}&season={season_str}'
            
            html = html_call(url)
            soup = BeautifulSoup(html,'html.parser')
            table_html=soup.find('div',{'id':'player_details_resultate'}).table
            
            df_score_table = pd.read_html(str(table_html))[0] # Returns list of all tables on 
            
            #tables = pd.read_html(url)
            #kvp = tables[-1]['Pt.'].iloc[:-1]
            kvp = df_score_table['Pt.'].iloc[:-1]
            
            if season == 20:
                seasons += [season_str] * 14
            else:
                seasons += [season_str] * 18
            weeks += list(kvp.keys())
            scores += list(kvp.values)
            
        except:
            print(sys.exc_info())
            if season == 20:
                week_count = 14
            else:
                week_count = 18
            seasons += [season_str] * week_count
            weeks += list(range(1,week_count+1))
            scores += [0] * week_count
        
        print(f'seasons: {len(seasons)}, weeks: {len(weeks)}, scores: {len(scores)}')
    return seasons, weeks, scores
```

```python
seasons, weeks, scores  = fetch_dream_score('278')
#print(f'seasons: {len(seasons)}, weeks: {len(weeks)}, scores: {len(scores)}')
```

```python
player_score_dict = { 'player_name':[],'season':[], 'week':[],'score':[]}
for (player,dream_id) in zip(df_social_media['player'],df_social_media['dream_id']):
    print(player)
    seasons, weeks, scores = fetch_dream_score(dream_id)
    player_score_dict['player_name'] += [player] * len(scores)
    player_score_dict['season'] += seasons
    player_score_dict['week'] += weeks
    player_score_dict['score'] += scores
    print(f'seasons: {len(seasons)}, weeks: {len(weeks)}, scores: {len(scores)},')
df_player_scores = pd.DataFrame(player_score_dict)
```

```python
df_player_scores.to_csv( '../data/player_scores.csv', index=False)
```
