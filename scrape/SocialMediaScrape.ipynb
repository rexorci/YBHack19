```python
import pandas as pd
import requests
```

```python
df_social_media = pd.read_csv('sm_accounts.csv')
```

```python
df_social_media
```

```python
twitter_url = 'https://cdn.syndication.twimg.com/widgets/followbutton/info.json?screen_names=Ackda'
```

```python
def fetch_twitter_follower_count(username):
    try:
        twitter_url = f'https://cdn.syndication.twimg.com/widgets/followbutton/info.json?screen_names={username}'
        resp = requests.get(url=twitter_url)
        data = resp.json() # Check the JSON Response Content documentation below
        return data[0]['followers_count']
    except:
        return 0
```

```python
fetch_twitter_follower_count('Ackda')
```

```python
def fetch_insta_follower_count(username):
    try:
        insta_url = f'https://www.instagram.com/{username}/?__a=1'
        resp_insta = requests.get(url=insta_url)
        data = resp_insta.json()
        follower_count = data['graphql']['user']['edge_followed_by']['count']
        return follower_count
    except:
        return 0
```

```python
df_social_media['insta_followers'] = df_social_media['instagram'].apply(lambda x: fetch_insta_follower_count(x))
```

```python
df_social_media['twitter_followers'] = df_social_media['twitter'].apply(lambda x: fetch_twitter_follower_count(x))
```
